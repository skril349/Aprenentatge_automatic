```python

def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[column] = np.clip(data[column], lower_bound, upper_bound)
numerical_columns = diabetes_df.select_dtypes(include=['int64', 'float64']).columns
for col in numerical_columns:
    remove_outliers_iqr(diabetes_df, col)

```

### **2. Reducció de Dimensionalitat (0.5 punts)**

En aquesta part de la PAC, utilitzarem **Anàlisi de Components Principals (PCA)** per millorar la informació dels pacients diabètics amb l'objectiu de facilitar la separació en el model de classificació. El nostre objectiu és garantir que el model PCA capturi patrons útils per distingir millor els pacients diabètics dels no diabètics.

**Riscos d'Overfitting i Com Evitar-los:**

Quan construïm el model PCA només amb pacients diabètics, pot existir el risc que els components principals capturin patrons específics d'aquest grup i no es generalitzin bé per als pacients no diabètics. Per mitigar aquest risc, seguirem un enfocament robust amb **validació creuada**, el qual:

1. **Mitjana el sobreajustament**, avaluant la estabilitat dels components principals a través de diferents particions de dades de pacients diabètics.
2. **Garanteix una millor generalització**, ja que seleccionem un conjunt estable de components mitjançant l'**averiguació de les càrregues** dels components principals a través dels diferents plegaments (folds) de la validació creuada.

### **Procés Recomanat**

1. **Preparació:** Es separen els pacients diabètics per a l'entrenament del PCA. Divideix les dades diabètiques en entrenament/validació (80%) i proves (20%), random_state=42.
2. **Normalització i Validació Creuada per al PCA:** Utilitzem validació creuada (n_splits=5, shuffle=True, random_state=42) per normalitzar les dades i avaluar la consistència dels components principals.
3. **Averiguació dels Components Principals:** Es fa la mitjana de les càrregues dels components principals per obtenir un conjunt estable.
4. **Projecció de les Dades Restants:** Utilitzem els components estables per projectar tant les dades de test dels pacients diabètics com les dades dels pacients no diabètics.
5. **Creació del Conjunt d'Entrenament per a Classificació:** El conjunt de dades projectat s'utilitza com a entrada per als models de classificació.

Aquest enfocament garanteix que el model PCA no sobreajusti als pacients diabètics utilitzats per construir-lo, i ens permet utilitzar descriptors coherents per a tot el conjunt de dades. Això millora la robustesa del model de classificació, minimitzant el risc de biaix i millorant la separació dels pacients en base a les seves característiques projectades.

- 2.5. Sobre l'emmagatzematge dels components principals

En treballar amb models de PCA utilitzant la llibreria `scikit-learn`, és important entendre que l'objecte del PCA (`pca_final`) **no guarda automàticament els components principals associats a les mostres transformades**. El PCA només manté els components del model (és a dir, les combinacions lineals de les variables originals) i la informació de la variància explicada.

Per això, si voleu conservar els **valors dels components principals (scores)** associats a cada mostra transformada per al seu ús futur, haureu de fer-ho **manualment**, creant un DataFrame separat. Això us permetrà:

1. **Reutilitzar les dades**: Podreu carregar-les directament en el futur sense haver de repetir el càlcul del PCA.
2. **Assegurar la consistència**: Eviteu diferències si el conjunt de dades original canvia o si s'afegeixen noves mostres.
3. **Preparar les dades per a l'anàlisi**: Podeu combinar els scores amb altres variables o utilitzar-los directament en models de classificació.

Consideracions finals:
- Recordeu que aquest procés s'ha de repetir per a qualsevol conjunt de dades que vulgueu projectar al model PCA (diabetic_test_data, non_diabetic_data).
- Aquesta separació us permet mantenir el model PCA enfocat únicament en transformacions i conservar els scores com a dades processades.

Aquesta particularitat pot semblar poc intuïtiva al principi, però assegura que el model es manté modular i fàcil d'utilitzar per a altres transformacions.

```python

def specificity_score(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    return tn / (tn + fp)

# models
models = {
    "Logistic Regression": LogisticRegression(random_state=42, max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42, n_estimators=100),
    "SVM": SVC(random_state=42, probability=True)
}
```

Per comprovar-ho, crearem la funció de comparació en la qual enviarem els resultas d'evaluació i els models i obtindrem així una comparació.

```python
def plot_model_comparison(evaluation_results, y_test, models, X_test):
    plt.figure(figsize=(8, 8))
    plt.title("ROC Curve Comparison")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.plot([0, 1], [0, 1], color="navy", lw=2, linestyle="--", label="Chance")

    for model_name, metrics in evaluation_results.items():
        if metrics["ROC-AUC"] != "N/A":
            y_probs = models[model_name].predict_proba(X_test)[:, 1]
            fpr, tpr, _ = roc_curve(y_test, y_probs)
            plt.plot(fpr, tpr, lw=2, label=f"{model_name} (AUC = {metrics['ROC-AUC']:.2f})")

    plt.legend(loc="lower right")
    plt.tight_layout()
    plt.show()

```

```python
plot_model_comparison(evaluation_results, y_test, models, X_test)
```

```python
def evaluate_models(X_train, X_test, y_train, y_test, models):
    """
    Funció per entrenar i avaluar models, retornant mètriques i traçant corbes ROC.
    Mostra el millor model basat en ROC-AUC.

    Paràmetres:
        X_train (DataFrame): Característiques d'entrenament.
        X_test (DataFrame): Característiques de prova.
        y_train (Series): Etiquetes d'entrenament.
        y_test (Series): Etiquetes de prova.
        models (dict): Diccionari amb noms dels models i els seus objectes inicialitzats.

    Retorna:
        dict: Mètriques d'avaluació per a cada model.
    """

    evaluation_results = {}
    best_model = None
    best_auc = 0

    for model_name, model in models.items():
        model.fit(X_train, y_train)

        y_pred = model.predict(X_test)
        y_probs = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None

        accuracy = accuracy_score(y_test, y_pred)
        sensitivity = recall_score(y_test, y_pred)
        specificity = specificity_score(y_test, y_pred)
        roc_auc = roc_auc_score(y_test, y_probs) if y_probs is not None else 0
        cm = confusion_matrix(y_test, y_pred)

        evaluation_results[model_name] = {
            "Accuracy": accuracy,
            "Sensitivity": sensitivity,
            "Specificity": specificity,
            "ROC-AUC": roc_auc,
            "Confusion Matrix": cm
        }

        if roc_auc > best_auc:
            best_auc = roc_auc
            best_model = model_name

    plt.figure(figsize=(6, 6))
    plt.title("ROC Curve Comparison")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.plot([0, 1], [0, 1], color="navy", lw=2, linestyle="--", label="Chance")

    for model_name, metrics in evaluation_results.items():
        if metrics["ROC-AUC"] > 0:
            y_probs = models[model_name].predict_proba(X_test)[:, 1]
            fpr, tpr, _ = roc_curve(y_test, y_probs)
            plt.plot(fpr, tpr, lw=2, label=f"{model_name} (AUC = {metrics['ROC-AUC']:.2f})")

    plt.legend(loc="lower right")
    plt.tight_layout()
    plt.show()

    print(f"El \033[1mmillor model\033[0m basat en el resultat de ROC-AUC es:  \033[1m{best_model}\033[0m amb AUC = \033[1m{best_auc:.2f}\033[0m")

    return evaluation_results


```

Per a aplicar l'ensamble learning us proposem seguir els següents passos:

---

**1. Classe `CascadingClassifier`**

###### **Propòsit:**
- Implementa una tècnica personalitzada d'aprenentatge ensemble on les prediccions d'un primer model (`base_model1`) s'afegeixen com a característiques addicionals a les dades d'entrada per a un segon model (`base_model2`).

###### **Components clau:**
- **Mètode `__init__`:**
  - Inicialitza el classificador amb dos models base.
- **Mètode `fit`:**
  - Entrena el `base_model1` amb les característiques d'entrada (`X`) i les etiquetes (`y`).
  - Genera les prediccions del `base_model1` i les afegeix com una nova columna a les característiques d'entrada.
  - Entrena el `base_model2` amb aquest conjunt de dades augmentat.
- **Mètode `predict`:**
  - Utilitza el `base_model1` entrenat per predir les etiquetes de les dades d'entrada.
  - Augmenta les característiques d'entrada amb aquestes prediccions i les passa al `base_model2` per obtenir les prediccions finals.
- **Mètode `predict_proba`:**
  - Similar a `predict`, però retorna les probabilitats de classe per a les dades d'entrada.

###### **Llibreries i tècniques clau:**
- **`np.hstack`:** Combina arrays per columnes per afegir les prediccions com a característiques.
- **`BaseEstimator` i `ClassifierMixin`:** Garanteixen la compatibilitat amb els pipelines i mètodes d'avaluació de Scikit-learn.

---

**2. Funció `evaluate_ensemble`**

###### **Propòsit:**
- Avalua el rendiment dels models ensemble utilitzant mètriques com l'accuracy, la sensibilitat, l'especificitat i el ROC-AUC. També genera una matriu de confusió i la corba ROC.

###### **Visualització:**
- Utilitza `RocCurveDisplay` per traçar la corba ROC.
- Millora la interpretabilitat imprimint mètriques i la matriu de confusió.

---

**3. Definició de models ensemble**

###### **Propòsit:**
- Defineix un conjunt de models ensemble per avaluar tant amb les característiques originals com amb les transformades pel PCA.

###### **Models inclosos:**
- **Bagging:** Combina múltiples estimadors entrenats amb subconjunts aleatoris de dades.
  - Exemple: `BaggingClassifier(estimator=DecisionTreeClassifier())`.
- **Boosting (AdaBoost):** Entrena seqüencialment estimadors, corregint errors anteriors.
  - Exemple: `AdaBoostClassifier(n_estimators=50,algorithm="SAMME", random_state=42)`.
- **Stacking:** Combina prediccions de múltiples models utilitzant un meta-model.
  - Exemple: `StackingClassifier(final_estimator=LogisticRegression(max_iter=500))`.
- **Cascading:** Utilitza la classe personalitzada `CascadingClassifier` descrita anteriorment.

###### **Característiques clau:**
- **Dues variants:**
  - **Característiques originals:** Models entrenats directament amb les dades originals.
  - **Característiques PCA:** Models entrenats amb dades transformades pel PCA.
- **Flexibilitat:** Permet afegir o modificar fàcilment models a `ensemble_models_original` i `ensemble_models_pca`.

**4. Models de Stacking per a ús explícit**
- Defineix explícitament dos models de stacking (`stacking_original` i `stacking_pca`) per estandarditzar-ne les configuracions:
  - **`estimators`:** Models base utilitzats per a les prediccions (per exemple, Decision Tree i Random Forest).

**Paràmetres clau:**
- **`max_iter`:** Incrementat a 500 per a Logistic Regression per garantir la convergència.
- **`cv`:** Especifica el nombre de particions per a cross-validation.

<strong>Implementació:</strong>
Aplica els passos anteriorment descrits per a obtenir els models i el testing de les tècniques de Baggin, Boosting, Stacking, Cascading, per a les característiques originals i les de principals components:

```python
# Definim la classe CascadingClassifier
class CascadingClassifier(BaseEstimator, ClassifierMixin):
    def __init__(self, base_model1, base_model2):
        self.base_model1 = base_model1  # Primer model
        self.base_model2 = base_model2  # Segon model

    def fit(self, X, y):
        # Entrenem el primer model
        self.base_model1.fit(X, y)

        # Prediccions del primer model
        pred1 = self.base_model1.predict(X).reshape(-1, 1)

        # Augmenta les característiques d'entrada
        X_augmented = np.hstack((X, pred1))

        # Entrenem el segon model
        self.base_model2.fit(X_augmented, y)
        return self

    def predict(self, X):
        # prediccions del primer model
        pred1 = self.base_model1.predict(X).reshape(-1, 1)
        X_augmented = np.hstack((X, pred1))

        # prediccions finals del segon model
        return self.base_model2.predict(X_augmented)

    def predict_proba(self, X):
        # probabilitats del primer model
        pred1 = self.base_model1.predict(X).reshape(-1, 1)
        X_augmented = np.hstack((X, pred1))
        # probabilitats del segon model
        return self.base_model2.predict_proba(X_augmented)

def evaluate_ensemble(X_train, X_test, y_train, y_test, models):
    results = {}
    plt.figure(figsize=(6, 6))
    plt.title("Comparació de la Corba ROC")
    plt.xlabel("Taxa de Falsos Positius")
    plt.ylabel("Taxa de Veritables Positius")
    plt.plot([0, 1], [0, 1], color="navy", lw=2, linestyle="--", label="Atzar")

    for name, model in models.items():
        # Entrena el model
        model.fit(X_train, y_train)

        # Genera prediccions
        y_pred = model.predict(X_test)
        y_probs = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None

        # Mètriques
        accuracy = accuracy_score(y_test, y_pred)
        sensitivity = recall_score(y_test, y_pred)
        specificity = specificity_score(y_test, y_pred)
        roc_auc = roc_auc_score(y_test, y_probs) if y_probs is not None else "N/A"
        cm = confusion_matrix(y_test, y_pred)

        results[name] = {
            "Accuracy": accuracy,
            "Sensitivity": sensitivity,
            "Specificity": specificity,
            "ROC-AUC": roc_auc,
            "Confusion Matrix": cm
        }

        # Corba ROC
        if roc_auc != "N/A":
            fpr, tpr, _ = roc_curve(y_test, y_probs)
            plt.plot(fpr, tpr, lw=2, label=f"{name} (AUC = {roc_auc:.2f})")

    plt.legend(loc="lower right")
    plt.tight_layout()
    plt.show()

    return results
```

### **5. Transfer Learning per a Dades Tabulars (2 punts)**

- 5.1. Què és el *Transfer Learning*?

El *Transfer Learning*, o aprenentatge per transferència, és una tècnica avançada dins del camp de la intel·ligència artificial i l'aprenentatge automàtic. La idea principal és senzilla però molt poderosa: aprofitar el coneixement adquirit per un model quan ha estat entrenat en una tasca per aplicar-lo a una altra tasca diferent però relacionada.

##### Metàfora per entendre-ho
Imagina que aprens a tocar el piano. Una vegada has après les bases, com ara llegir partitures o coordinar les mans, aprendre a tocar un altre instrument, com el violí, serà més fàcil perquè moltes habilitats que has adquirit ja són aplicables. Això és exactament el que fa el *Transfer Learning* en l'àmbit dels models d'aprenentatge profund.

---

5.1.1. Com funciona?
1. **Model preentrenat**:
   - En el *Transfer Learning*, es comença amb un model ja entrenat en una gran quantitat de dades generals (com imatges de milers de categories diferents).
   - Aquest model ha après patrons generals (per exemple, identificar línies, formes i textures en imatges).

2. **Transferència de coneixement**:
   - Es reutilitzen les primeres capes del model preentrenat, que contenen aquest coneixement general.
   - Només cal ajustar (o afegir) les capes finals perquè el model pugui especialitzar-se en la nova tasca específica.

3. **Entrenament amb dades específiques**:
   - Amb un conjunt més petit de dades específiques de la tasca (com imatges de pacients amb o sense diabetis), el model s'entrena per refinar el coneixement i adaptar-se.

---

5.1.2. Per què utilitzar el *Transfer Learning*?
1. **Estalvi de temps i recursos**:
   - No cal entrenar un model des de zero, ja que el model preentrenat ja ha realitzat gran part de la feina d’aprenentatge.
   - Això redueix significativament el temps i els costos computacionals.

2. **Rendiment més alt amb dades limitades**:
   - Si només disposem d’un conjunt petit de dades específiques, el *Transfer Learning* ajuda a evitar problemes com el sobreajustament (*overfitting*), aprofitant el coneixement previ.

3. **Resultats robustos**:
   - Els models preentrenats ja han estat provats amb grans conjunts de dades, la qual cosa garanteix una base sòlida i generalitzable per treballar.

---

##### Exemple pràctic
Suposem que volem classificar si una imatge mostra un pacient amb diabetis o no. Podem utilitzar un model preentrenat com ResNet50 (que ha après a reconèixer patrons generals d’imatges) i adaptar-lo a la nostra tasca, sense necessitat d’entrenar-lo des de zero.

El *Transfer Learning* permet enfocar-nos en la nostra tasca específica sense perdre temps construint i entrenant un model completament nou. És com pujar a l'esquena d'un gegant per arribar més lluny!

```python
def transform_to_images(data, input_shape):
    n_samples = data.shape[0]
    n_features = np.prod(input_shape)
    if data.shape[1] != n_features:
        raise ValueError(f"Number of features ({data.shape[1]}) does not match target shape ({input_shape}).")
    return data.reshape((n_samples, *input_shape))
```

- 5.4. **Conversió de dades a float32**

**Per què cal convertir les dades a `float32`?**

En l’àmbit de l’aprenentatge automàtic i el processament de dades, és habitual que sigui necessari convertir les dades al format `float32`. Pros:

---

**Eficiència de memòria**
- **Reducció de l’ús de memòria**: El format `float32` utilitza 4 bytes de memòria, mentre que `float64` (el format predeterminat en moltes llibreries com NumPy o Pandas) utilitza 8 bytes. En conjunts de dades grans, aquesta diferència pot reduir significativament el consum de memòria.
- **Adaptabilitat per a GPUs**: Les GPUs solen treballar amb `float32` de manera nativa. Utilitzar aquest format evita un ús innecessari de memòria i millora l'eficiència computacional.

---

**Optimització del rendiment**
- **Càlculs més ràpids**: Els càlculs amb `float32` són més ràpids que amb `float64`, especialment en GPUs, ja que moltes eines d’aprenentatge profund (com TensorFlow o PyTorch) estan optimitzades per a `float32`.
- **Temps de transferència reduït**: Transferir dades entre CPU i GPU és més ràpid quan les dades estan en format `float32`, gràcies al menor volum de dades.

---

**Evitar problemes de compatibilitat**
- **Requisits de les eines**: Moltes eines d’aprenentatge automàtic, com TensorFlow o Keras, esperen que les dades d’entrada estiguin en format `float32`. Si s’utilitza `float64`, es poden generar advertències o errors que obligaran a fer la conversió posteriorment.
- **Models preentrenats**: La majoria dels models preentrenats i operacions en aprenentatge profund estan dissenyats per treballar amb `float32`. Això assegura la compatibilitat i evita desajustos entre tipus de dades.

---

**Impacte mínim en la precisió**
- **Precisió suficient**: Per a la majoria de tasques d’aprenentatge automàtic, la precisió que ofereix `float32` és més que suficient. Tot i que `float64` proporciona una precisió més alta, aquesta normalment no és necessària en aplicacions pràctiques.
- **Control d’errors**: El format `float32` ajuda a mitigar problemes com desbordaments o subdesbordaments de valors, que poden ser més probables amb `float64`.

---

**Bones pràctiques en la indústria**
- **Consistència**: Mantenir el mateix tipus de dades en tota la pipeline (entrada, pesos del model, sortides) evita comportaments inesperats o incompatibilitats durant els càlculs.
- **Estàndard**: La majoria de llibreries i eines d’aprenentatge automàtic utilitzen `float32` com a format predeterminat. Això fa que aquest format sigui un estàndard per a garantir la compatibilitat.

---

**Quan no és necessari convertir a `float32`?**
- **Requisits de precisió alta**: Si l'aplicació requereix càlculs molt precisos (per exemple, simulacions científiques), podria ser més adequat utilitzar `float64`.
- **Conjunts de dades petits**: Si el conjunt de dades és petit i el consum de memòria no és un problema, la conversió pot no ser imprescindible.

---

Per tant, a l’hora de preparar les dades per a una pipeline d’aprenentatge automàtic, **heu de convertir-les a `float32` per millorar el rendiment, reduir el consum de memòria i garantir la compatibilitat amb les eines modernes.** Això és especialment important si esteu treballant amb conjunts de dades grans o entrenant models en GPUs.

- X_test: Testing data (already reshaped).

<strong>Implementació:</strong>
Crea la funció per a una CNN personalitzada per a les dades tabulars remodelades en petites estructures semblants a imatges en 2D.

```python
def create_custom_cnn(input_shape):
    """
    Crea una CNN personalitzada per a dades tabulars reestructurades en petites estructures similars a imatges.

    Arguments:
        input_shape (tuple): Forma de les dades d'entrada (p. ex., (2, 4, 1)).

    Retorna:
        Sequential: Model CNN compilat.
    """

    model = Sequential([
        Conv2D(32, kernel_size=(2, 2), activation='relu', input_shape=input_shape, padding='same'),
        BatchNormalization(),
        Conv2D(64, kernel_size=(2, 2), activation='relu', padding='same'),
        BatchNormalization(),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.5),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model
```

<div class="alert alert-block alert-info">
<strong>Implementació:</strong>
Crea la funciò per a entrenar i avaluar un model CNN personalitzat per a dades tabulars remodelades com a imatges.
</div>

```python
def train_evaluate_custom_cnn(X_train, X_test, y_train, y_test, input_shape, dataset_type="original"):
    """
    Entrena i avalua una CNN personalitzada per a dades tabulars reestructurades en petites estructures similars a imatges.

    Arguments:
        X_train (numpy.ndarray): Dades d'entrenament (reestructurades).
        X_test (numpy.ndarray): Dades de prova (reestructurades).
        y_train (numpy.ndarray): Etiquetes d'entrenament.
        y_test (numpy.ndarray): Etiquetes de prova.
        input_shape (tuple): Forma de les dades d'entrada (p. ex., (7, 1, 1)).
        dataset_type (str): "original" o "pca" per indicar el tipus de conjunt de dades.

    Retorna:
        dict: Mètriques incloent precisió, sensibilitat, especificitat i ROC-AUC.
    """

    # model CNN
    model = create_custom_cnn(input_shape)

    # Train
    history = model.fit(X_train, y_train, epochs=20, batch_size=16, validation_data=(X_test, y_test), verbose=1)

    # evaluacio
    y_pred = (model.predict(X_test) > 0.5).astype(int)
    y_probs = model.predict(X_test).flatten()

    accuracy = accuracy_score(y_test, y_pred)
    sensitivity = recall_score(y_test, y_pred)
    specificity = recall_score(y_test, y_pred, pos_label=0)
    roc_auc = roc_auc_score(y_test, y_probs)
    conf_matrix = confusion_matrix(y_test, y_pred)

    # mètriques
    print(f"\nDataset Type: {dataset_type}")
    print(f"Accuracy: {accuracy:.2f}")
    print(f"Sensitivity: {sensitivity:.2f}")
    print(f"Specificity: {specificity:.2f}")
    print(f"ROC-AUC: {roc_auc:.2f}")
    print("Confusion Matrix:")
    print(conf_matrix)

    fpr, tpr, _ = roc_curve(y_test, y_probs)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc:.2f})', color='darkorange')
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve ({dataset_type.capitalize()} Dataset)')
    plt.legend(loc="lower right")
    plt.show()

    return {"Accuracy": accuracy, "Sensitivity": sensitivity, "Specificity": specificity, "ROC-AUC": roc_auc}
```

## **6. Optimització del Model i Sintonització d'Hiperparàmetres (2 punts)**

- 6.1. Sintonització d'hiperparàmetres amb GridSearchCV per al millor model d'ensemble





<div class="alert alert-block alert-info">
<strong>Implementació:</strong>
Seleccioneu el model d'ensemble amb millor rendiment (per exemple, Stacking o Bagging). Definiu una graella d'hiperparàmetres rellevants. Per exemple: Per a Bagging: Nombre d'estimadors (n_estimators), mida màxima de mostres (max_samples); Per a Stacking: Models base i paràmetres associats als classificadors meta Utilitzeu GridSearchCV per explorar les combinacions possibles dels hiperparàmetres i seleccionar la millor configuració.
</div>

```python
def optimize_stacking(X_train, y_train):
    """
    Optimitza StackingClassifier utilitzant GridSearchCV.

    Arguments:
        X_train (numpy.ndarray): Dades d'entrenament.
        y_train (numpy.ndarray): Etiquetes d'entrenament.

    Retorna:
        El millor estimador i els resultats de GridSearch.
    """

    # Definimmodels base
    base_estimators = [
        ('dt', DecisionTreeClassifier(random_state=42)),
        ('rf', RandomForestClassifier(random_state=42))
    ]

    # Definim el classificador meta
    final_estimator = LogisticRegression(max_iter=500)

    # StackingClassifier
    stacking_model = StackingClassifier(
        estimators=base_estimators,
        final_estimator=final_estimator
    )

    # hiperparàmetres
    param_grid = {
        'final_estimator__C': [0.1, 1.0, 10.0],
        'final_estimator__solver': ['lbfgs', 'liblinear'],
        'cv': [3, 5]
    }

    # Configuració GridSearchCV
    grid_search = GridSearchCV(
        estimator=stacking_model,
        param_grid=param_grid,
        scoring='accuracy',
        cv=5,
        verbose=2,
        n_jobs=-1
    )

    grid_search.fit(X_train, y_train)

    return grid_search.best_estimator_, grid_search.best_params_
```

```python
best_stacking_model, best_stacking_params = optimize_stacking(X_pca_train, y_pca_train)
print("Best Stacking Model:", best_stacking_model)
print("Best Parameters for Stacking:", best_stacking_params)


```

- 6.2. Avaluació:

<div class="alert alert-block alert-info">
<strong>Implementació:</strong>
Apliqueu el model amb els millors hiperparàmetres al conjunt de test. Compareu les mètriques del model optimitzat amb el model inicial.

**Consideracions finals**
- La sintonització d’hiperparàmetres sovint proporciona millores significatives, però pot ser computacionalment costosa. Prioritzeu els paràmetres més influents.
- Les mètriques d’avaluació del conjunt de test són claus per determinar si l’optimització ha millorat el model de manera generalitzada.
- Documenteu els hiperparàmetres seleccionats, els canvis en les mètriques i qualsevol observació rellevant per justificar les vostres decisions.

```python
def evaluate_optimized_model(best_model, X_test, y_test, initial_metrics):
    """
    Avalua el model optimitzat amb el conjunt de dades de prova i compara amb les mètriques inicials.

    Arguments:
        best_model: El model optimitzat (després de GridSearchCV).
        X_test: Dades de prova.
        y_test: Etiquetes de prova.
        initial_metrics: Mètriques del model inicial abans de l'ajust de paràmetres.

    Retorna:
        dict: Mètriques del model optimitzat.
    """
    # Prediccions i probabilitats
    y_pred = best_model.predict(X_test)
    y_probs = best_model.predict_proba(X_test)[:, 1]

    # Càlcul de mètriques
    accuracy = accuracy_score(y_test, y_pred)
    sensitivity = recall_score(y_test, y_pred)
    specificity = recall_score(y_test, y_pred, pos_label=0)
    roc_auc = roc_auc_score(y_test, y_probs)
    conf_matrix = confusion_matrix(y_test, y_pred)

    # mètriques
    print("\n=== Mètriques del Model Optimitzat ===")
    print(f"Precisió: {accuracy:.2f}")
    print(f"Sensibilitat: {sensitivity:.2f}")
    print(f"Especificitat: {specificity:.2f}")
    print(f"ROC-AUC: {roc_auc:.2f}")
    print("Matriu de Confusió:")
    print(conf_matrix)

    # Comparaió
    print("\n=== Comparació amb el Model Inicial ===")
    print(f"{'Mètrica':<15}{'Model Inicial':<20}{'Model Optimitzat'}")
    print("-" * 50)
    for metric, value in initial_metrics.items():
        optimized_value = {
            "Accuracy": accuracy,
            "Sensitivity": sensitivity,
            "Specificity": specificity,
            "ROC-AUC": roc_auc
        }.get(metric, None)
        print(f"{metric:<15}{value:<20.2f}{optimized_value:.2f}")


    return {
        "Accuracy": accuracy,
        "Sensitivity": sensitivity,
        "Specificity": specificity,
        "ROC-AUC": roc_auc
    }
```

Addicionalment és important remarcar:

**Importància de la Sensibilitat i l'Especificitat en Dades Mèdiques:**

En el context mèdic, la **sensibilitat** i l'**especificitat** són mètriques crucials que mesuren la capacitat d'un test diagnòstic per identificar correctament els casos positius i negatius, respectivament. Aquestes mètriques són especialment importants perquè ajuden a prevenir errors diagnòstics que podrien portar a tractaments incorrectes o mancats de tractament.

- **Sensibilitat** (també coneguda com a taxa de veritable positiu): Mesura la proporció de positius reals que són correctament identificats pel model. Una alta sensibilitat és crucial en condicions mèdiques on no detectar una malaltia (un fals negatiu) podria tenir conseqüències greus per al pacient. Per exemple, en el diagnòstic de càncer, una alta sensibilitat és essencial per assegurar que gairebé tots els pacients amb la malaltia són identificats per a un tractament precoç.

- **Especificitat** (també coneguda com a taxa de veritable negatiu): Mesura la proporció de negatius reals que són correctament identificats. Una alta especificitat és important per evitar falsos positius, que podrien portar a pacients sense la malaltia a sotmetre's a procediments invasius o estressants innecessaris. Per exemple, en el cribratge de malalties cardíaques, una alta especificitat redueix el nombre de pacients que rebran diagnòstics erronis de condicions que no tenen.

**Relació entre Sensibilitat i Especificitat:**

La relació entre sensibilitat i especificitat sovint s'involucra en un **compromís**; optimitzar un model per millorar la sensibilitat generalment reduirà la seva especificitat i viceversa. Això es deu a que l'augment de la sensibilitat implica abaixar el llindar per classificar un resultat com a positiu, el que incrementa el nombre de falsos positius i disminueix l'especificitat. A la inversa, augmentar l'especificitat solament augmenta el llindar, el que pot portar a més falsos negatius i disminuir la sensibilitat.

**Aplicacions en el Món Real:**

La selecció d'un model en un entorn mèdic depèn de la malaltia específica i del context del cribratge o diagnòstic. En malalties on les conseqüències de no detectar un cas positiu són greus, com en el càncer o infeccions greus, es pot preferir un model amb alta sensibilitat. En canvi, en condicions on un diagnòstic fals positiu podria portar a tractaments innecessaris o perjudicials, es preferirà un model amb alta especificitat.

La comprensió d'aquest equilibri i la seva correcta aplicació són essencials per desenvolupar eines diagnòstiques que no només són tècnicament competents sinó també èticament responsables, assegurant que els pacients reben el diagnòstic més precis possible amb les mínimes repercussions negatives.
